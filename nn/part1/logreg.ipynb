```python
import numpy as np
np.random.seed(1)
%matplotlib inline
```

```python
points = []
colors = []
n=100

for _ in range(n):
    point = np.random.random(2) * 5.5 + 0.25
    points.append(point)
    colors.append(1 if -point[0] + 6 > point[1] else 0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticRegr = LogisticRegression()
logisticRegr.fit(points, colors)
```

```python
 w1, w2 = logisticRegr.coef_[0]
 b = logisticRegr.intercept_[0]
```

```python
import matplotlib.pyplot as plt

color = ["r", "b"]

for i in range(n):
    plt.scatter([points[i][0]], [points[i][1]], color=color[colors[i]])

func = lambda x: (-1/w2) * (b + w1*x)
x = np.linspace(0, 6)
plt.plot(x, list(map(func, x)))
```
