```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image = Image.open("img/attention_scores.png")
image = image.convert("RGB")
pixels = np.array(image.getdata())
pixels = pixels.reshape((image.height, image.width, 3))

for i in range(15):
    for j in range(14):
        h = (i + 0.5) * image.height / 15
        w = (j + 0.5) * image.width / 14
        h = int(h)
        w = int(w)

        # Assign a 10x10 pixel block to the average of the 100 pixels
        block = pixels[h:h+10, w:w+10]
        avg = np.mean(block, axis=(0, 1))
        pixels[h:h+10, w:w+10] = [255, 0, 0]

plt.imshow(pixels)
```
